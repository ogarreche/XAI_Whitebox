#### Here starts the process to converting the feature scores to sparsity values.

```python
with open(output_file_name, "a") as f:print('XAI: ', XAI_name ,'', file = f)
print('XAI: ', XAI_name )

print('---------------------------------------------------------------------------------')
print('Generating Sparsity Graph')
print('---------------------------------------------------------------------------------')
print('')
for i in range(0, 3):
    feature_val = features[i]
    # Find the minimum and maximum values in the list
    min_value = min(feature_val)
    max_value = max(feature_val)

    # Normalize the list to the range [0, 1]
    normalized_list = []
    for x in feature_val:
        if max_value - min_value == 0:
            normalized_list.append(0)
        else:
            normalized_list.append((x - min_value) / (max_value - min_value))
    count_below_threshold = 0
    Spar = []
    # print('Sparsity = ',Sparsity)
    X_axis = []
    #----------------------------------------------------------------------------
    for j in range(0, 11):
        j/10
        threshold = j/10
        for value in normalized_list:
            if value <= threshold:
                count_below_threshold += 1

        Sparsity = count_below_threshold/len(normalized_list)
        Spar.append(Sparsity)
        X_axis.append(j/10)
        count_below_threshold = 0


    #---------------------------------------------------------------------------
    
    with open(output_file_name, "a") as f:print(order[i] ,'', file = f)
    print(order[i])
    with open(output_file_name, "a") as f:print('y_axis = ', Spar ,'', file = f)
    print('y_axis = ', Spar )
    

with open(output_file_name, "a") as f:print('x_axis = ', X_axis ,'', file = f)
print('x_axis = ', X_axis )

```
